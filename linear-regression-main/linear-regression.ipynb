```python
x = pd.DataFrame(my_data, columns = ["production_budget_usd"])
y = pd.DataFrame(my_data, columns = ['worldwide_gross_usd'])

my_data.describe()
```

```python
pltt.scatter(x, y, alpha = 0.3)

pltt.xlabel("Budget")
pltt.ylabel("Revenue")

pltt.xlim(0, 4.250000e+08)
pltt.ylim(0, 2.783919e+09)
my_data
```

```python
reg = LinearRegression()
my_data
```

```python
reg.fit(x, y)
my_data
```

```python
reg.coef_
reg.intercept_
my_data
```

```python
pltt.scatter(x, y, alpha = 0.3)

pltt.xlabel("Budget")
pltt.ylabel("Revenue")

pltt.plot(x, reg.predict(x), color='red')

pltt.xlim(0, 4.250000e+08)
pltt.ylim(0, 2.783919e+09)
```
